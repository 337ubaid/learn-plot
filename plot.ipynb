```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('healthexp')
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df, hue='Country')
```

```python
# what country has the highest life expectancy?
sns.boxplot(data=df, x='Country', y='Life_Expectancy')
```

```python
# is life expectancy correlated with Spending USD?
sns.scatterplot(data=df, x='Life_Expectancy', y='Spending_USD', hue='Country')
```

```python
# make a heatmap for each year and each country
df_pivot = df.pivot_table(index='Country', columns='Year', values='Life_Expectancy')
sns.heatmap(df_pivot, cmap='viridis')
```
